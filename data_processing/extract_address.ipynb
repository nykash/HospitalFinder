```python
for f, nam in zip(files, ["acute_stroke", "cesarian", "complications_safety", "patient_satisfaction", "vaginal"]):
    df_temp = pd.read_csv(f)
    mean_value = df_temp[df_temp["Entity Name"] == "New York State Average"].iloc[0]["Score"]
    df_temp = df_temp[df_temp["Entity Name"] != "New York State Average"].drop_duplicates()
    scores = []
    significance = []


    for i in range(df.shape[0]):
        row = df.iloc[i]
        row_temp = df_temp[df_temp["PFI"] == row["Facility.Id"]]
        if row_temp.shape[0] == 0:
            row_temp = df_temp[df_temp["PFI"] == str(row["Facility.Id"])]
            if row_temp.shape[0] == 0:
                scores.append("NA")
                significance.append("NA")
                continue

        row_temp = row_temp.iloc[0]

        scores.append(row_temp["Score"])
        significance.append(row_temp["Significance"])

    df[f"{nam}_stat"] = scores
    df[f"{nam}_sig"] = significance
    df[f"{nam}_mean"] = [mean_value]*len(scores)
```
